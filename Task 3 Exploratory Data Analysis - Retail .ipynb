#### 1.  Visualizing Segment of the Consumer buying from the SuperStore.

```python
retail['Segment'].value_counts()  #Counts the number of Particular Customer Segments Present in the Dataset.
```

```python
plt1.plot(kind='bar',color='green')  # It Plots the information obtaied by normalizing our segment values.
```

```python
#Pie Chart
plt1.plot(kind='pie',figsize=(6,8))
plt.legend()
plt.title('Segment')
```

##### Conclusion: It can be observed that more than 50% of the Customer segment is of consumers,then corporate and then home office,So the Store owner should priortize accordingly

#### 5.  Analysing the Profits as per the Customer Segments in the Superstore.

```python
fig,ax=plt.subplots()
colors={'Consumer':"red","Corporate":'Cyan',"Home Office":"Green"}
ax.scatter(x=retail['Sales'],y=retail['Profit'],c=retail['Segment'].apply(lambda x: colors[x]))   #Scatter plot of different 
                                                                                                  #customer segments
plt.xlabel("Sales") 
plt.ylabel("Profit")
plt.title("Sales and Profit Analysis on all the Segments")
plt.show()
```

#### Conclusion: Here we can derive from this Scatter plot that most of the profit comes from the cusumer segment, so they should focus of satisfying as many needs of consumer segment as possible.

#### 6.  Profit and Discount Relationship using frequency plot in order to determine Discount amount Maximizing Profit.

```python
retail1=retail.loc[(retail['Segment']=='Consumer')&(retail['Discount']==0.1)]    #If Discount is 10%
retail1
```

```python
retail1['Profit'].plot.hist(bins=50)
```

```python
retail2=retail.loc[(retail['Segment']=='Consumer')&(retail['Discount']==0.2)]    #If Discount is 20%
retail2['Profit'].plot.hist(bins=50,color='green')
```

```python
retail3=retail.loc[(retail['Segment']=='Consumer')&(retail['Discount']==0.5)]    #If Discount is 50%
retail3['Profit'].plot.hist(bins=50,color='blue')
```

##### Conclusion: We can easily see that on plotting profit with various values of discounts,max,profit obtained is at discount value about 10%.So Store owner can't really afford to give discount more than 10% to the consumer.

#### 7.  Overall average profit mapping with the segment given the different values of Discount.

```python
retail4=retail.groupby(['Segment','Discount']).Profit.median()
retail4
```

```python
retail4.plot(kind='bar',stacked=True,figsize=(10,10),ylabel='Profit')
```

#### Conclusion: Here average profit in all the segments along with possible discount values is given to have a better understanding of the discount segment relation.

#### 8.  States where maximum Sales have occurred for the Superstore in United States.

```python
#Seaborn Library is used to plot this figure.

plt.figure(figsize=(10,6))
states=retail['State'].value_counts()[0:20]  #Get the count of particular states mentioned in whole column and sort in descending.
sns.barplot(x=states ,y=states.index , palette='deep')
plt.title("States Instances to get a fait idea of which states have high density present in the dataset")
plt.xlabel("Number of Instances of States in the whole dataset.")
plt.ylabel("States of United States")
```

##### Conclusion : Here the figure is self explanatory where one can determine in which states maximum products are sold and top 20 states out of them are depicated here.

#### 9.  Identifying the Sub-Category where Maximum profit is obtained.

```python
#Top 10 Profit making sub-categories are shown here in the plot.

retail.groupby("Sub-Category")['Profit'].max().nlargest(10).plot.bar()
```

##### Conclusion: Sub-Categories like copiers,binders etc, generate more profit than others, so owner should really focus on selling and maintaining inventory of these.

#### 12. Overall Sales data in the US according to the sub-category in all the Segments as whole.

```python
flip_xlabels = theme(axis_text_x = element_text(angle=90, hjust=1),figure_size=(10,5),
                     axis_ticks_length_major=10,axis_ticks_length_minor=5)
(ggplot(retail, aes(x='Sub-Category', fill='Sales')) + geom_bar() + facet_wrap(['Segment']) 
+ flip_xlabels +theme(axis_text_x = element_text(size=12))+ggtitle("Sales From Every Segment Of United States of Whole Data"))
```

##### Conclusion: Here we can get overall sales idea of all the sub categories segment wise,It can be considered as an overview of our dataset
